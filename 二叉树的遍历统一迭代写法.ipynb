```python
from typing_extensions import Deque
from collections import deque
class BiTreeNode:
  def __init__(self, val):
    self.val = val
    self.lchild = None
    self.rchild = None


a = BiTreeNode('1')
b = BiTreeNode('2')
c = BiTreeNode('3')
d = BiTreeNode('4')
e = BiTreeNode('5')
f = BiTreeNode('6')
g = BiTreeNode('7')
h = BiTreeNode('8')
i = BiTreeNode('9')

b.lchild = c
b.rchild = e
c.lchild = f
c.rchild = g
g.lchild = a
g.rchild = h
e.lchild = i
e.rchild = d

root = b

class BinaryTree:
  def __init__(self, root):
    self.root = root

  def pre_order(self, root):
    if not root:
      return

    print(root.val, end=',')
    self.pre_order(root.lchild)
    self.pre_order(root.rchild)

  def in_order(self, root):
    if not root:
      return

    self.in_order(root.lchild)
    print(root.val, end=',')
    self.in_order(root.rchild)

  def post_order(self, root):
    if not root:
      return

    self.post_order(root.lchild)
    self.post_order(root.rchild)
    print(root.val, end=',')

  def level_order(self, root): #BFS
    q = deque()
    q.append(root)
    print(root.val, end=',')
    while len(q)>0:
      curNode = q.popleft()
      if curNode.lchild:
        print(curNode.lchild.val, end=',')
        q.append(curNode.lchild)

      if curNode.rchild:
        print(curNode.rchild.val, end=',')
        q.append(curNode.rchild)
    return


tree = BinaryTree(root)
tree.pre_order(root)
print("")
tree.in_order(root)
print("")
tree.post_order(root)
print("")
tree.level_order(root)

```
